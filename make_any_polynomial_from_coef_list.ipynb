```python
# Function any poly returns a polynomal function with degree = len(coefficients)-1

import matplotlib.pyplot as plt

def any_poly(coefs):
    return lambda x: sum([coefs[i] * x**i for i in range(len(coefs))]) 

#test
coefs=[0.,0.,0.,0.,0.,0.]
deltas=[0.,1.,-0.05,0.001,-0.0001,0.000001]

for i in range(5):
    plt.plot([(any_poly(coefs)(x)) for x in range(21)])
    coefs = [sum(x) for x in zip(coefs,deltas)] # increment coefs by delta
    
plt.show()
                   
                   
```
